Linear Regration직선에서 각점들이 얼마나 떨어져 있는지 오차를 그래프로 표시

```
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

print('X shape:', x.shape)
print(x)
```

```
sns.scatterplot(x=x, y=y, color='g')
```

```
sns.regplot(x=x, y=y, color='g')
```

# lowess
- 가중치 W를 감안하여 그린다
- locally weighted robust scatterplot smoothing의 약자

```
sns.regplot(x=x, y=y,  color='g', lowess=True) 
```

```
sns.residplot(x=x, y=y, lowess=True, color='g') # 오차가 양의방향 또는 음의방향으로 편향되어있는지 알수있다
```
