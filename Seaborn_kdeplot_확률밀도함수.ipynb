```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
rs = np.random.RandomState(50)
```

```
iris = sns.load_dataset('iris')
print(iris.shape)
iris.head()
```

```
setosa = iris.query("species == 'setosa'")
virginica = iris.query("species =='virginica'")

print(setosa.shape)
print(virginica.shape)
```

```
setosa.describe()
```

```
sns.kdeplot(setosa['sepal_length'])
sns.kdeplot(virginica['sepal_length'])
```

```
sns.kdeplot(setosa['sepal_length'], setosa['sepal_width'])
```

```
# shade - 색상을 칠한다
sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], shade=True)
```

```

sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], shade=True)
sns.scatterplot(data=setosa, x='sepal_width', y = 'sepal_length')
```

```
f, ax = plt.subplots(figsize=(8,8))
ax.set_aspect('equal') # 그래프의 가로, 세로간격을 동일하게 만들어준다

sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], 
            shade=True, shade_lowest=False)
sns.kdeplot(virginica['sepal_width'], virginica['sepal_length'], 
            shade=True, shade_lowest=False)
```

```
f, ax = plt.subplots(figsize=(8,8))
ax.set_aspect('equal') # 그래프의 가로, 세로간격을 동일하게 만들어준다

sns.kdeplot(setosa['sepal_width'], setosa['sepal_length'], 
            shade=True, shade_lowest=False, 
            cmap='Blues')
sns.kdeplot(virginica['sepal_width'], virginica['sepal_length'], 
            shade=True, shade_lowest=False, 
            cmap='Reds')

ax.text(2.5, 8.2,'Virginica', size=16, color='blue')
ax.text(3.8, 4.5,'Setosa', size=16, color='red')
```
