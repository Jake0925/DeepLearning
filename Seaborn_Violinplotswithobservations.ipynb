```
import numpy as np
import seaborn as sns
import pandas as pd
```

```
sns.set(style='whitegrid') # seaborn 설정
```

```
rs = np.random.RandomState(0)
n, p = 40,8
d = rs.normal(0,2, (n, p))
d += np.log(np.arange(1, p+1)) * -5 +10
pd.DataFrame(d)
```

```
pal = sns.cubehelix_palette(p, rot=-.5, dark=.3)
sns.palplot(pal) # pal수치값을 palplot을통해 이미지로 표현
```

```
sns.violinplot(data=d, palette=pal, inner='points') # 값이 음수로 갈수록 색상이 진해진다
```
