```python
import numpy as np
import tensorflow as tf
```

```python
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]
```

image.png

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1)) # units = output shape, input_dim = input shape
sgd = tf.keras.optimizers.SGD(lr=0.1) # SGD == standard gradient descendent 최소비용이 발생하도록 찾아가는 함 , lr은 learning rate
tf.model.compile(loss='mse', optimizer=sgd) # mse == mean_squared_error, 1/m * sig (y'-y)^2
tf.model.summary()
tf.model.fit(x_train, y_train, epochs=200) # 모델 훈련
```

```python
y_predict = tf.model.predict(np.array([5, 4])) # 예측값
print(y_predict)
```
