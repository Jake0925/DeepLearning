```
import pandas as pd
import seaborn as sns
sns.set_theme(style="whitegrid", palette="muted")
```

```
iris = sns.load_dataset('iris')
iris
```

```
iris.info()
```

```
iris = pd.melt(iris, 'species', var_name = 'measurement')
iris
```

scatter Plot으로 표현한경우

```
sns.scatterplot(x='measurement', y = 'value', hue = 'species', data=iris)
```

swarmplot

```
sns.swarmplot(x='measurement', y='value', hue='species', palette=['r','c', 'y'], data=iris)
```

violinplot

```
tips = sns.load_dataset('tips')
tips.sample()
```

```
sns.violinplot(x='day', y='total_bill', hue='smoker',  split=True, data=tips)
sns.swarmplot(x='day', y='total_bill', hue='smoker', split=True, data=tips, color='red')
```
