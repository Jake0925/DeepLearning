```
import seaborn as sns
sns.set_theme(style="white")
df=sns.load_dataset('iris')
```

```
g = sns.PairGrid(df, diag_sharey=False) # iris의 항목별 그래프틀 생성
g.map(sns.scatterplot)
```

```
g = sns.PairGrid(df, diag_sharey=False, hue='species') # iris의 항목별 그래프틀 생성
g.map(sns.scatterplot)
```

```
g = sns.PairGrid(df, diag_sharey=False, hue='species') # iris의 항목별 그래프틀 생성
g.map(sns.scatterplot)
g.map_diag(sns.kdeplot)
```

```
g = sns.PairGrid(df, diag_sharey=False, hue='species') # iris의 항목별 그래프틀 생성
g.map_lower(sns.kdeplot) # 대각선을 기준 아래쪽엔 kdeplot를 그린다
g.map_upper(sns.scatterplot) # 대각선을 기준 위쪽엔 scatterplot를 그린다
g.map_diag(sns.kdeplot)
```
