```
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the example tips dataset
tips = sns.load_dataset("tips")
tips
```

```
tips.info()
```

```
tips.sample() # 데이터중 랜덤하게 한 데이터를 가져온다
```

```
sns.violinplot(x='day', y='total_bill', data=tips) # x축은 범주형데이터 y축엔 수치형데이터 삽입
```

```
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```
# order 그래프에 표시할 항목순서를 조절
sns.violinplot(x='time', y='tip', data=tips, order=["Dinner", "Lunch"])
```

```
tips["weekend"] = tips["day"].isin(["Sat", "Sun"]) # day항목중 Sat와 Sun을 포함하면 'weekend'항목을 추가하여 안에 True 아니편 False값 입력
tips.sample()
```

```
sns.violinplot(x='day', y='total_bill', data=tips, hue='weekend', dodge=False) # weekend의 값이  True/False에 따라 다른색상으로 표시
sns.despine(left=True) # 투두리 제거
```

```
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1, # inner의 설정에 따라 내부의 막대표시변경, ex)point. box, quartile(사분위수표시) 
               palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```
