```
import seaborn as sns
import numpy as np
sns.set(style='ticks')
dots = sns.load_dataset('dots')
```

```
print(dots.shape)
dots.head(10)
```

```
dots.info()
```

```
dots.coherence.unique() # dots.coherence의 값은 수치형이미만 6가지로만 정해져있다
```

```
ChangeColor_palette = dict(zip(dots.coherence.unique(), sns.color_palette('rocket_r', 6)))
ChangeColor_palette
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate")
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence')
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence', kind='line')
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence',kind='line',
    col='align') # col에 의해 align값에 따라 새로운 그래프 생성
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence',kind='line',
    col='align', size='choice') 
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence',kind='line',
    col='align', size='choice', size_order=["T1", "T2"], legend='full') 
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence',kind='line',
    col='align', size='choice', size_order=["T1", "T2"],
    legend='full',
    height=5, aspect=.8, facet_kws=dict(sharex=False), # 그래프 크기 변경 heignt - 세로, aspect - 가로, facet_kws=dict(sharex=False)- x축 범위를 값의 시작부터 끝의 값까지만 표시
    palette = ChangeColor_palette)
```

```
sns.relplot(data=dots,
    x="time", y="firing_rate", hue='coherence',
    col='align', size='choice', size_order=["T1", "T2"],
    legend='full',
    height=5, aspect=.8, facet_kws=dict(sharex=False), # 그래프 크기 변경 heignt - 세로, aspect - 가로, facet_kws=dict(sharex=False)- x축 범위를 값의 시작부터 끝의 값까지만 표시
    palette = ChangeColor_palette)
```
