```
import seaborn as sns
sns.set_theme(style="ticks")
df = sns.load_dataset("anscombe")
df
```

```
df.describe()
```

```
print(df['dataset'] == 'I') # 데이터중 dataset열의 값이'I'에 해당하는지 확인

print('\n', df[df['dataset'] == 'I'].describe()) # 'i'에 해당하는 행만 추출
```

```
df[df['dataset'] == 'II'].describe()
```

```
df[df['dataset']=='III'].describe()
```

상관관계체크
- 4개 그룹모두 상관관계가 거의 비슷하다

```
print('I', df[df['dataset'] == 'I'].corr()) 
print('\nII',df[df['dataset'] == 'II'].corr())
print('\nIII',df[df['dataset'] == 'III'].corr())
print('\nIV',df[df['dataset'] == 'IV'].corr())
```

```
sns.lmplot(x='x', y='y', data=df)
```

```
sns.lmplot(x='x',y='y', data=df, hue='dataset') # hue - 항목중 'dataset'항목을 기준으로 분류
```

```
# regplot은 lmplot와 유사하다
# regplot는 hue속성을 사용할수 없다, lmplot와 가장큰 차이점 

sns.regplot(x='x',y='y', data=df) 
```

```
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset')
```

```
# ci - 신뢰구간설정
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci=None)
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci=95) # 95%를 포함함 신뢰구간표시
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci=50) # 50%를 포함한 신뢰구간표시
```

```
# col_wrap속성
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci=None, col_wrap=2)
```

```
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci=None, col_wrap=2, palette='muted',height=3)
```

```
# alpha - 투명도, s- 점들의 크기변경
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})

sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 200, "alpha": 0.1})
```
