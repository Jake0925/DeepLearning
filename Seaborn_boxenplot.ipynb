```
import seaborn as sns
import pandas as pd
sns.set_theme(style="whitegrid")
```

```
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```
diamonds.info()
```

### Box안의 가운데 수평선은 중앙값을 나타낸다

```
sns.boxplot(data=diamonds, x = 'clarity', y = 'carat') # 수용구간을 알아보기 힘들다
```

```
sns.boxenplot(data=diamonds, x = 'clarity', y = 'carat')
```

```
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.boxenplot(data=diamonds, x = 'clarity', y = 'carat',
              order = clarity_ranking)
```

```
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.boxenplot(data=diamonds, x = 'clarity', y = 'carat',
              order = clarity_ranking,
              scale ='linear')
```

```
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.boxenplot(data=diamonds, x = 'clarity', y = 'carat',
              order = clarity_ranking,
              scale ='linear',
              color='b')
```

```
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.violinplot(data=diamonds, x = 'clarity', y = 'carat',
              order = clarity_ranking,
              color='b')
```
